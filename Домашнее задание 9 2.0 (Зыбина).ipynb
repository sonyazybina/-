```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import statsmodels.formula.api as smf
```

# Задача 1.1 sleep equation #1

```python
df = pd.read_csv('https://raw.githubusercontent.com/artamonoff/Econometrica/master/python-notebooks/data-csv/sleep75.csv')
df
```

```python
model = smf.ols(data=df, formula='sleep~totwrk+age+male+smsa+south').fit()
hat_y = model.predict()
```

```python
hat_y
```

```python
df['slp2'] = hat_y**2/3600
df['slp3'] = hat_y**3/216000
df['slp4'] = hat_y**4/12960000
df
```

```python
model2 = smf.ols(data=df, formula='sleep~totwrk+age+male+smsa+south+slp2+slp3+slp4').fit()
model2.params
```

```python
R2_un = model2.rsquared
R2_re = model.rsquared
```

```python
F_obs = (R2_un - R2_re)/(1-R2_un)*(len(df)-9)/3
F_obs.round(3)
```

```python
alpha=0.05
stats.f.ppf(1-alpha,3,len(df)-6).round(3)
```

$ F_{obs} < F_{cr} \Rightarrow H_0$ не отвеграем
